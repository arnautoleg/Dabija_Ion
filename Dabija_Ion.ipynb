# Categorical graphs

```python
categorical_variables = df.select_dtypes(include='category')
type(categorical_variables)
print(categorical_variables)
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from statsmodels.stats.proportion import proportion_confint


for el in categorical_variables:
  # Assuming df is your DataFrame with two categorical variables, 'Comportament Manager' and 'Profesia'
  # Create a contingency table
  contingency_table = pd.crosstab(categorical_variables['Lotul'], categorical_variables[el])


  # Define the desired order for variables

  variable1_order = ['Control', 'Studiu']  # Replace with your actual category names

  # Convert variables to CategoricalDtype with specified order
  contingency_table.index = pd.Categorical(contingency_table.index, categories=variable1_order, ordered=True)



  # Sort the index and columns based on the specified order
  contingency_table.sort_index(axis=0, inplace=True)
  contingency_table.sort_index(axis=1, inplace=True)
  contingency_table

  # Perform the Chi-squared test
  chi2, p, _, _ = chi2_contingency(contingency_table)

  # Print the Chi-squared statistic and p-value
  # print(f"Chi-squared: {chi2}")
  # print(f"P-value: {p}")

  # Calculate column percentages
  col_percentages = contingency_table.div(contingency_table.sum(axis=1), axis=0) * 100

  # Create a new DataFrame for formatted annotations
  annot_data = pd.DataFrame(index=col_percentages.index, columns=col_percentages.columns)
  for i in range(contingency_table.shape[0]):
      for j in range(contingency_table.shape[1]):
          annot_data.iloc[i, j] = f"{col_percentages.iloc[i, j]:.1f}%"

  # Calculate exact confidence intervals for each cell
  confidence_intervals = pd.DataFrame(index=col_percentages.index, columns=col_percentages.columns)
  for i in range(contingency_table.shape[0]):
      for j in range(contingency_table.shape[1]):
          data = contingency_table.iloc[:, j].values
          ci = proportion_confint(data[i], nobs=sum(data), method='beta')
          confidence_intervals.iloc[i, j] = (ci[0] * 100, ci[1] * 100)

  # Increase the size of the plot
  plt.figure(figsize=(10, 6))

  # Plot the heatmap with formatted annotations
  # 'coolwarm', 'viridis', 'BuPu', YlGnBu
  sns.heatmap(col_percentages, annot=annot_data, fmt="", cmap="coolwarm")

  # Add Chi-squared statistic and p-value as text annotations
  plt.text(0.5, -0.5, f"Chi-squared: {chi2:.2f}", ha='center', va='center', fontsize=12, color='black')
  if p < 0.001:
    plt.text(0.5, -0.7, f"P-value: p < 0.001", ha='center', va='center', fontsize=12, color='black')
  else:
    plt.text(0.5, -0.7, f"P-value: {p:.4f}", ha='center', va='center', fontsize=12, color='black')

  # Annotate each cell with exact confidence interval
  for i in range(contingency_table.shape[0]):
     for j in range(contingency_table.shape[1]):
         plt.text(j + 0.5, i + 0.8, f"CI {confidence_intervals.iloc[i, j][0]:.1f}-{confidence_intervals.iloc[i, j][1]:.1f}", ha='center', va='top', fontsize=10, color='black')

  # Add the chi-squared test result to the plot
  #plt.text(0.5, 0.5, f"Chi-Squared Statistic: {chi2:.2f}\nP-value: {p:.4f}",
  #         fontsize=12, ha="center", va="center", transform=plt.gca().transAxes)

  # Rotate y-axis tick labels to a horizontal position
  plt.yticks(rotation=0)
  plt.title(f'Asocierea Lotul și {el}. Frecvente relative cu 95%CI')
  plt.show()
```

# Numerical graphs

```python
numerical_variables = df.select_dtypes(include='float64')
#print(numerical_variables)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu

def boxplot_with_mannwhitney(df, group_column, numeric_column):
    """
    Create a boxplot and perform a Mann-Whitney U test for two groups in a DataFrame.

    Parameters:
        df (DataFrame): The input DataFrame.
        group_column (str): The column containing group labels.
        numeric_column (str): The numeric column for comparison.

    Returns:
        None (displays the plot).
    """
    # Print descriptive statistics for the numeric column grouped by the specified group column
    print(round(df.groupby([group_column])[numeric_column].describe(), 1))
    print()

    # Create a boxplot
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=group_column, y=numeric_column, data=df)
    plt.title('Boxplot cu testul Mann-Whitney U')

    # Perform Mann-Whitney U test
    group1 = df[df[group_column] == df[group_column].unique()[0]][numeric_column]
    group2 = df[df[group_column] == df[group_column].unique()[1]][numeric_column]
    stat, p_value = mannwhitneyu(group1, group2, alternative='two-sided')

    # Get the maximum value in the numeric column for dynamic positioning
    max_value = max(df[numeric_column])

    # Add Mann-Whitney U test results as text annotation with dynamic positioning
    if p_value < 0.001:
        plt.text(0.4, max_value + 0.1 * max_value, f'Mann-Whitney U Test: U={stat:.1f}, p-value<0.001', ha='center', va='bottom', color='black')
    else:
        plt.text(0.4, max_value + 0.1 * max_value, f'Mann-Whitney U Test: U={stat:.1f}, p-value={p_value:.4f}', ha='center', va='bottom', color='black')


    low_limit = 0
    if el in ['Stabilitatea primara',  'Stabilitatea 6 luni']:
      low_limit = -8


    plt.ylim(low_limit, max_value + 0.2 * max_value)  # Adjust the ylim for better visualization
    plt.show()
```

```python
for el in numerical_variables:
  boxplot_with_mannwhitney(df, 'Lotul', el)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import friedmanchisquare
from scikit_posthocs import posthoc_nemenyi_friedman
import scikit_posthocs as sp
import numpy as np


def friedman_and_heatmap(data, category_col, numerical_col, category_order):
    # Set the style of the visualization
    sns.set(style="whitegrid")

    # Create a box plot
    plt.figure(figsize=(15, 6))
    plt.subplot(1, 2, 1)
    sns.boxplot(x=category_col, y=numerical_col, data=data, palette='Set2', order=category_order)
    plt.xlabel(category_col)
    plt.ylabel(numerical_col)
    plt.ylim(0, max(data[numerical_col]) + max(data[numerical_col])*0.2)
    plt.title('Box Plot pentru IO in dinamica')

    # Calculate Friedman test statistic
    f_statistic, p_value = friedmanchisquare(*[group[1][numerical_col] for group in data.groupby(category_col)])

    # Display Friedman test statistics
    print(f"Friedman test: Chi-square = {f_statistic}, p-value = {p_value}")

    # Perform post-hoc Nemenyi test for pairwise comparisons


    data = np.array([list(df[df['Lotul'] == 'Control']['IOR preoperator, mm']),
                     list(df[df['Lotul'] == 'Control']['IO postoperator, mm']),
                     list(df[df['Lotul'] == 'Control']['IO 6 luni, mm'])])

    #perform Nemenyi post-hoc test
    nemenyi_df = sp.posthoc_nemenyi_friedman(data.T)
    nemenyi_df.columns = ['IOR preoperator, mm', 'IO postoperator, mm', 'IO 6 luni, mm']
    nemenyi_df.index = ['IOR preoperator, mm', 'IO postoperator, mm', 'IO 6 luni, mm']

    plt.subplot(1, 2, 2)
    sns.heatmap(nemenyi_df, cmap='viridis', annot=True, linewidths=.5)
    plt.title('Evaluarea comparativa IO in dinamica. Lotul de control')


    print(nemenyi_df)

     # Adjust layout for better spacing
    plt.tight_layout()

    plt.show()


```

# Paired test control

```python
df_long = pd.melt(df[df.Lotul  =='Control'], value_vars=['Stabilitatea primara', 'Stabilitatea 6 luni'],
                   var_name='Dinamica', value_name='Stabilitatea')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import wilcoxon



def boxplot_with_wilcoxon(df, group_column, numeric_column):
    """
    Create a boxplot and perform a Wilcoxon signed-rank test for paired samples in a DataFrame.

    Parameters:
        df (DataFrame): The input DataFrame.
        group_column (str): The column containing group labels.
        numeric_column (str): The numeric column for comparison.

    Returns:
        None (displays the plot).
    """
    # Print descriptive statistics for the numeric column grouped by the specified group column
    print(round(df.groupby([group_column])[numeric_column].describe(), 1))
    print()

    # Create a boxplot
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=group_column, y=numeric_column, data=df)
    plt.title('Boxplot cu testul Wilcoxonpentru masurari repetate')

    # Perform Wilcoxon signed-rank test
    group1 = df[df[group_column] == df[group_column].unique()[0]][numeric_column]
    group2 = df[df[group_column] == df[group_column].unique()[1]][numeric_column]
    stat, p_value = wilcoxon(group1, group2)

    # Get the maximum value in the numeric column for dynamic positioning
    max_value = max(df[numeric_column])

    # Add Wilcoxon signed-rank test results as text annotation with dynamic positioning
    if p_value < 0.001:
        plt.text(0.4, max_value + 0.1 * max_value, f'Wilcoxon Test: W={stat:.1f}, p-value<0.001', ha='center', va='bottom', color='black')
    else:
        plt.text(0.4, max_value + 0.1 * max_value, f'Wilcoxon Test: W={stat:.1f}, p-value={p_value:.4f}', ha='center', va='bottom', color='black')

    plt.ylim(-8, max_value + 0.2 * max_value)  # Adjust the ylim for better visualization
    plt.show()
```

```python
boxplot_with_wilcoxon(df_long, 'Dinamica', 'Stabilitatea')
```
